```
import numpy as np
import matplotlib.pyplot as plt

a=b=np.arange(0,3,0.02)
c=np.exp(a)
d=c[::-1]

plt.plot(a,c,"b--", linewidth=3, label="oh yes")
plt.plot(a,d,"r-.", label="good")
plt.plot(a,c+d,"y.", label="total")
plt.legend()
```

```
import numpy as np
import matplotlib.pyplot as plt
N=5
mmeans=(20,35,30,35,27)
wmeans=(25,32,33,21,26)
id=np.arange(N)
print(id)
width=0.33
p1=plt.bar(id, mmeans,width, color='g')
p2=plt.bar(id+0.33, wmeans,width, color='r')
plt.xlabel("people")
plt.ylabel("means")
plt.xticks(id+width/2,("G1","G2","G3","G4","G5"))
plt.yticks(np.arange(0,50,5))
plt.title("means by gender")
plt.legend((p1[0],p2[0]),("men","woman"))
plt.show()
```
