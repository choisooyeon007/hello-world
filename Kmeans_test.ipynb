```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```
df=pd.read_csv("covid 27Jul (3).csv")
df.head()
```

```
df1=df.set_index('dong')
df1.head()
```

```
df2=df1.dropna(axis=1)
df2.head()
```

```
#df2.columns.dtype
#df2.values.dtype
#df2.index.dtype
#df2.index.astype('object')
df2.dtypes
```

```
kmeans=KMeans(n_clusters=4).fit(df2.values)
kmeans.cluster_centers_
```

```
kmeans.labels_
df2['cluster']=kmeans.labels_
df2.head()
```

```
df00=df2[df2['cluster']==0]
df01=df2[df2['cluster']==1]
df02=df2[df2['cluster']==2]
df03=df2[df2['cluster']==3]
df00.shape
df01.shape
df02.shape
df03.shape
df03.head()
```

```
plt.figure(figsize=(20,20))

plt.subplot(4,1,1)
plt.ylim(-100,300)
plt.title('group 0')
plt.plot(df00) 

plt.subplot(4,1,2)
plt.ylim(-100,300)
plt.title('group 1')
plt.plot(df01) 

plt.subplot(4,1,3)
plt.ylim(-100,300)
plt.title('group 2')
plt.plot(df02) 

plt.subplot(4,1,4)
plt.ylim(-100,300)
plt.title('group 3')
plt.plot(df03) 


plt.show()

```
