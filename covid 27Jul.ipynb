```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
df=pd.read_csv("unix (4).csv", )
df.head()
```

```
df1=df.set_index('dong')
df1.head()
```

```
df2=np.transpose(df1)
df2.head()
```

```
df2.columns
```

```
plt.figure(figsize=(20,10))
plt.ylim(-100,300)
plt.plot(df2) 
plt.title('COVID')
plt.xlabel('datetime')
plt.ylabel('change')
plt.legend(df2.columns, title='dong', bbox_to_anchor=(0.5,-0.05), loc='upper center', ncol=10)
plt.tight_layout()
plt.show()

```
