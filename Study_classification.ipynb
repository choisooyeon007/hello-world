```
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.DESCR)
```

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df = pd.DataFrame(iris.data, columns=iris.feature_names )
#print(df)
sy = pd.Series(iris.target, dtype="category")
#print(sy)
sy=sy.cat.rename_categories(iris.target_names)
#print(sy)
df['species']=sy
df.head()
```

```
sns.pairplot(df,hue='species')
plt.show()
```

```
sns.distplot(df[df.species == "setosa"]["petal length (cm)"], rug=True, label="setosa")
sns.distplot(df[df.species != "setosa"]["petal length (cm)"], rug=True, label="others")
plt.legend()
plt.show()
```

```
sns.distplot(df[df.species == "virginica"]["petal length (cm)"], hist=True, rug=True, label="virginica")
sns.distplot(df[df.species == "versicolor"]["petal length (cm)"], hist=True, rug=True, label="versicolor")
plt.legend()
plt.show()
```

```
from sklearn.datasets import load_wine
wine = load_wine()
print(wine.DESCR)
```

```
df=pd.DataFrame(wine.data, columns=wine.feature_names)
dt=pd.Series(wine.target,dtype="category").cat.rename_categories(wine.target_names)
df['class']=dt
df.head()
```

```
sns.pairplot(df,vars=['alcohol', 'malic_acid', 'ash', 'flavanoids'], hue='class')
plt.show()
```

```
sns.distplot(df[dt == "class_0"]["flavanoids"], hist=True, rug=True, label='class_0') 
sns.distplot(df[dt == "class_2"]["flavanoids"], hist=True, rug=True,label='class_2')
plt.legend()
plt.show()
```

```
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.DESCR)
```

```
np.random.seed(0)
N = 5
M = 10
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0.1, wspace=0.05)
for i in range(N):
    for j in range(M):
        k = i*M+j
        ax = fig.add_subplot(N, M, k+1)
        ax.imshow(digits.images[k], cmap=plt.cm.Blues,interpolation="none" )
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title(digits.target_names[digits.target[k]])
plt.tight_layout()
plt.show() 
print(digits.target_names[digits.target[25]])     
```

```
from sklearn.datasets import fetch_olivetti_faces
olivetti = fetch_olivetti_faces()
print(olivetti.DESCR)
```

```

N = 2
M = 5
fig = plt.figure(figsize=(8, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
klist = np.random.choice(range(len(olivetti.data)), N * M)
#print(klist)
for i in range(N):
    for j in range(M):
        k = klist[i*M+j]
        ax = fig.add_subplot(N, M, i*M+j+1)
        ax.imshow(olivetti.images[k], cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title(olivetti.target[k])
plt.tight_layout()
plt.show()
        
```

```
np.random.seed(0)
N = 2
M = 5
fig = plt.figure(figsize=(8, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0.01, wspace=0.05)
klist = np.random.choice(range(len(lfw_pairs.data)), M)
for j in range(M):
    k = klist[j]
    ax1 = fig.add_subplot(N, M, j+1)
    ax1.imshow(lfw_pairs.pairs[k][0], cmap=plt.cm.summer)
    ax1.grid(False)
    ax1.xaxis.set_ticks([])
    ax1.yaxis.set_ticks([])
    plt.title(lfw_pairs.target_names[lfw_pairs.target[k]])
    ax2 = fig.add_subplot(N, M, j+1 + M)
    ax2.imshow(lfw_pairs.pairs[k][1], cmap=plt.cm.bone)
    ax2.grid(False)
    ax2.xaxis.set_ticks([])
    ax2.yaxis.set_ticks([])
plt.tight_layout()
plt.show()
```
