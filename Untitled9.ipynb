```
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

iris = pd.read_csv('iris.data', header=None, names=['sl', 'sw', 'pl', 'pw', 'cl'])
X = iris[['sw','sl']].to_numpy()
names, Y = np.unique(iris['cl'], return_inverse=True) # setosa, virginicia eta => 0,1,2

tree1 = DecisionTreeClassifier(max_depth=10, random_state=0).fit(X,Y)
tree2 = BaggingClassifier(DecisionTreeClassifier(max_depth=10), n_estimators=100, random_state=0).fit(X,Y)

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
  # Pick some colors for the iris'
myse,myvi,myve = (0.2,0.2,0.8),(0.2,0.8,0.2),(0.8,0.2,0.2)
cmap = [myse, myve, myvi] # based on the names array
x,y = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))

plt.subplot(121)
pred = tree1.predict(np.column_stack([x.reshape(-1), y.reshape(-1)]))
z1 = np.array([cmap[p] for p in pred]).reshape((200,200,3))
  # Put the result into a color plot
plt.imshow(z1, extent=(x_min,x_max,y_min,y_max), origin='bottom-left', aspect='auto')
  # Plot also the training points
X0 = X[Y==0]; X1 = X[Y==1]; X2 = X[Y==2]
_=plt.scatter(X0[:,0], X0[:,1], c=[myse], edgecolor='k',label='setosa')
_=plt.scatter(X1[:,0], X1[:,1], c=[myve], edgecolor='k',label='versicolor')
_=plt.scatter(X2[:,0], X2[:,1], c=[myvi], edgecolor='k',label='virginica')
plt.xlabel('Sepal width (cm)'); plt.ylabel('Sepal length (cm)'); plt.legend()

plt.title("Tree Model")

plt.subplot(122)
pred = tree2.predict(np.column_stack([x.reshape(-1), y.reshape(-1)]))
z2 = np.array([cmap[p] for p in pred]).reshape((200,200,3))
  # Put the result into a color plot
plt.imshow(z2, extent=(x_min,x_max,y_min,y_max), origin='bottom-left', aspect='auto')
  # Plot also the training points
X0 = X[Y==0]; X1 = X[Y==1]; X2 = X[Y==2]
_=plt.scatter(X0[:,0], X0[:,1], c=[myse], edgecolor='k',label='setosa')
_=plt.scatter(X1[:,0], X1[:,1], c=[myve], edgecolor='k',label='versicolor')
_=plt.scatter(X2[:,0], X2[:,1], c=[myvi], edgecolor='k',label='virginica')
plt.xlabel('Sepal width (cm)'); plt.ylabel('Sepal length (cm)'); plt.legend()
plt.title("Bagging Model")

plt.tight_layout()
plt.show()

```
